```python
%matplotlib notebook
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def randrange(n, vmin, vmax):
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure(figsize=(4,4))

ax = fig.add_subplot(111,projection='3d')
n = 25
xs = randrange(n, 5, 15)
ys = randrange(n, 10, 15)
zs = randrange(n, 10, 15)

yg = ax.scatter(xs, ys, zs, marker='o')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
plt.show()
```
