## Introduction to Numerical Computing with NumPy 

Numpy means data, crunshing a lot of data. And at it's heart is numpay array
It's not like a regulalr c++ or java array. Its more like a vector/matrix
Since data analysis is heavily based on arrays and matrix. Numpy can do algebraic operations and is optimized for super speeds while doing these operations on arrays/matrix
Has in-built matrix operations like product, inverse, determinant, linear systems , etc.

**numpy stands for numerical python**

Lets take an example. Lets say you have an array of some people weights and heights and you have to calculate BMI (body mass index)

```python
heights = [ 1.88, 1.70, 1.65, 1.55, 2.10]
weights = [ 70, 80, 50, 65, 98]

print(heights)
print(weights)
```

Can we do this ?

```python
weights / heights ** 2
```

A way to this is by having a for loop that goes throught each member and does calculations on them

```python
results = []
for h, w in zip(heights, weights):
    results.append(round(w / h ** 2 , 2))
print(results)
```

**I want you to improve this function below as homework assignment. Make it as short as possible**

```python
result = map(lambda item: round(item[0]/item[1] **2, 2),  zip(weights, heights))

for item in result:
    print(item)
```
