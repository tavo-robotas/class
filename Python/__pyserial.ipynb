In this lecture you will learn something cool. How to connect a physical device that sends data to python. How to simply visualise or interact with that data on python

First of all we need a device that send us that data. In this particular case we are going to use ultrasonic sensor <b>HC-SR04</b> and a small MCU (microcontroler unit) <b>Arduino Pro micro</b> but any MCU is suitable as long as it has enough pins to connect the ultrasonic device



This small Arduino MCU has to execute a small bit of code to get ultrasonic sensor data that is measuring distance from obstacle and send it over SPI (Serial protocol interface). This code is written in C but its quite simple and easy to understand, in case you don't understand it doesn't matter for now. Since this course is about python not C just make a copy into your arduino IDE and load it to MCU  

```python
#include <Ultrasonic.h>

Ultrasonic atstumoMatuoklis(3, 2); // Tigger , Emmiter 
int atstumas;

void setup() {
  Serial.begin(9600);
}

void loop() {
 
  atstumas = atstumoMatuoklis.read();
  Serial.println(atstumas);
  delay(1000); // 1000 ms = 1s , tai bus laiko intervalas kas kiek bus siuncami duomenys
}
```

Now the fun part. You have to install two libraries to able to read and visualize this data

```python
# install PySerial library , more info https://pyserial.readthedocs.io/en/latest/pyserial.html
pip install pyserial

# install vpython library, more info https://vpython.org/
pip install vpython
```

```python
import serial as com
from vpython import *


# labai svarbu atkreipti demesi i tai per koki com* yra prijungtas mikrokompiuteris.
# Kiekvienas USB kanalas turi skirtinga adresa. Kad suzinoti per kuri yra prijungtas jusu irenginys, tai gali paziureti
# per Arduino IDE > Tools > Port

data = com.serial_for_url('com4', baudrate=9600, timeout=1)
deze = box(pos=default_loc, size=vector(40, 20, 10), color=color.yellow)
pavadinimas = label(pos=vector(-150, 100, 0), height=20)

while 1:
    rate(40)
    if data.inWaiting() > 0:
        message = data.readline().decode('ascii')
        deze.pos.x = int(message)
        pavadinimas.text = 'Atstumas yra: ' + message
```
