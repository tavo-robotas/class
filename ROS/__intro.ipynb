## Terminologija

### Patys svarbiausi:

- Mazgas <code>**node**</code>. Tai mažiausias vykdomas procesas. Gali būti laikoma atskira vykdoma programa. ROS sistema susideda iš daugybės mazgų. Kiekvienas mazgas siunčia ir priima duomenys per pranešimų komunikacija nuo to priklauso ar mazgas yra skelbėjas<code>**publisher**</code>,ar prenumeratorius <code>**subscriber**</code>, gali būti abu.

- Pakuotė <code>**package**</code>. Vienas ar daugiau mazgų, kartu su mazgų vykdymo informacija. Grupė pakuočių kurios turi bendrą tikslą tai pat vadinamos metapakuotėmis

- Pranešimas <code>**message**</code>. Duomenys yra siunčiami ir priimami tarp mazgų pranešimai. Pranešimas gali būti skirtingo <code>**duomenų tipo**</code>, int, float, boolean. Tai pat gali būti struktūrizuotos kaip paprastos duomenų struktūros, masyvai kurie laiko žinutes kitose žinutėse.

## plepys.cpp

````python
#include "ros/ros.h"
#include "std_msgs/String.h"
#include <sstream>

int main(int argc, char **argv)
{
      ros::init(argc, argv, "kalbetojas");
      ros::NodeHandle mano_mazgas;
      ros::Publisher kalbetojas = mano_mazgas.advertise<std_msgs::String>("plepalai", 1000);
      ros::Rate loop_rate(1.0);
      int count = 0;
      while(ros::ok())
      {
            std_msgs::String msg;
            std::stringstream ss;
            ss << "Labas" << count;
            msg.data = ss.str();
            ROS_INFO("[plepys] pranesa %s\n", msg.data.c_str());
            kalbetojas.publish(msg);
            ros::spinOnce();
            loop_rate.sleep();
            count++;
      }
      return 0;
}
````

## klausytojas.cpp

````python
#include "ros/ros.h"
#include "std_msgs/String.h"

void plepalu_callback(const std_msgs::String::ConstPtr &pranesimas)
{
    ROS_INFO("[klausytojas] girdejo: [%s]\n", pranesimas->data.c_str());
}

int main(int argc, char **argv)
{
    ros::init(argc, argv, "klausytojas");
    ros::NodeHandle mazgas;
    ros::Subscriber klausytojas = mazgas.subscribe("plepalai", 1000, plepalu_callback);
    ros::spin();
    return 0;
}
````

````python
#!/usr/bin/env python
import rospy
from std_msgs.msg import String 
    def plepys():
        pub = rospy.Publisher('plepalai', String, queue_size=10)
        rospy.init_node('talker', anonymous=True)
        rate = rospy.Rate(10) # 10hz
        while not rospy.is_shutdown():
            hello_str = "Labas %s" % rospy.get_time()
            rospy.loginfo(hello_str)
            pub.publish(hello_str)
            rate.sleep()

if __name__ == '__main__':
    try:
        plepys()
    except rospy.ROSInterruptException:
        pass
````

````python
#!/usr/bin/env python
import rospy
from std_msgs.msg import String 

def callback(data):
    rospy.loginfo(rospy.get_caller_id() + "Girdėjau %s", data.data)     

def listener():
    rospy.init_node('listener', anonymous=True) 
    rospy.Subscriber("plepalai", String, callback)
    rospy.spin()
 
if __name__ == '__main__':
    listener()
````

````python
#!/usb/bin/env/python

import rospy
from ros_class.msg import Sensorius

import random

pub = rospy.Publisher('klimatas', Sensorius, queue_size=10)
rospy.init_node('sensoriaus_skelbejas', anonymous=True)
rate = rospy.Rate(1)
i = 0
while not rospy.is_shutdown():
    sensor = Sensor()
    sensor.id = 1
    sensor.name = 'bevardis'
    sensor.temperature = 24 + (random.random()*2)
    sensor.humidity = 33 + (random.random()*2)
    rospy.loginfo('Skelbiu:')
    rospy.loginfo(sensor)
    pub.publish(sensor)
    rate.sleep()
    i += 1
````

````python
import sys
import rospy

from ros_class.srv import Arithmetica
from ros_class.srv import ArithmeticaRequest
from ros_class.srv import ArithmeticaResponse


def launch_client(x, y):
      rospy.wait_for_service('sudek_du_skaicius')
      try:
            # sudek yra kliento objektas
            sudek = rospy.ServiceProxy('sudek_du_skaicius', Arithmetica)
            # atsakas yra atsakymo iš serverio objektas
            atsakas = sudek(x, y)
            return atsakas.sum
      except rospy.ServiceException, e:
            print("nepavyko susisiekti su serveriu: %s" % e)


def usage():
      return "%s [x, y]" % sys.argv[0]


if if __name__ == "__main__":
    if len(sys.argv) == 3:

          x = int(sys.argv[1])
          y = int(sys.argv[2]
      else:
            print(usage())
            sys.exit(1)
      print("Siunciu uzklausa %s + %s" % (x, y))
      print("%s + %s = %s" % (x, y, launch_client(x, y)))
````

<code>**C++**</code> versija

kuriame savo faila su plėtiniu <code>**.cpp**</code> src papkeje

````python
#include "ros/ros.h"
#include "ros_class/Arithmetica.h"
#include <cstdlib>

int main(int argc, char **argv)
{
      ros::init(argc, argv, "sudeties_klientas");
      if (argc != 3)
      {
            ROS_INFO("reikia argumentu: x, y");
            return 1;
      }

      ros::NodeHandle n;
      ros::ServiceClient klientas = n.serviceClient<ros_class::Arithmetica>("sudek_du_skaicius");
      ros_class::Arithmetica srv;
      srv.request.a = atoll(argv[1]);
      srv.request.b = atoll(argv[2]);

      if (klientas.call(srv))
      {
            ROS_INFO("Suma: %ld", (long int)srv.response.suma);
      }
      else
      {
            ROS_ERROR("Nepavyko pakviesti serviso sudek_du_skaicius");
            return 1;
      }

      return 0;
}
````

Padarykime <code>**vienkryptį**</code> judėjimą:

````python
#include "ros/ros.h"
#include "geometry_msgs/Twist.h"

ros::Publisher velocity_publisher;

void move(double speed, double distance, bool isForward);

int main(int argc, char **argv)
{
    ros::init(argc, argv, "turtlesim_cleaner")
    ros::NodeHandle n;
    
    velocity_publisher = n.advertise<geometry_msgs::Twist>("turtlesim/cmd_vel", 10);
    #move(2.0, 5.0, 1);
    ros::spin();
    return 0;
}

void move(double speed , double distance, bool isForward)
{
    geometry_msgs::Twist vel_msg;
    vel_msg.linear.x = isForward ? abs(speed) : -abs(speed); 
    vel_msg.linear.y = 0;
    vel_msg.linear.z = 0;
    
    vel_msg.angular.x = 0;
    vel_msg.angular.y = 0;
    vel_msg.angular.z = 0;
    
    double t0 = ros::Time::now().toSec();
    double current_distance = 0;
    ros::Rate loop_rate(10);
    do{
        velocity_publisher.publish(vel_msg);
        double t1 = ros::Time::now().toSec();
        current_distance = speed * (t1 - t0); # px/s
        
        ros::spinOnce();
        loop_rate.sleep();

    }while (current_distance <= distance);
    
    //stop the cleaner 
    vel_msg.linear.x = 0;
    velocity_publisher.publish(vel_msg);
}

````

````python
void rotate(double angular_speed, double relative_angle, bool clockwise)
{
    geometry_msgs::Twist vel_msg;
    vel_msg.linear.x = 0;
    vel_msg.linear.y = 0;
    vel_msg.linear.z = 0;
    
    vel_msg.angular.x = 0;
    vel_msg.angular.y = 0;
    #čia kaip (if - else) sąlyga, bet naudojant ternary operatorių
    vel_msg.angular.z = clockwise ? -abs(angular_speed) : abs(angular_speed);
    
    double current_angle = 0.0;
    double t0 = ros::Time::now().toSec();
    rps::Rate loop_rate(10);
    do{
        # judėjimas kampu matuojamas rad/s
        double t1 = ros::Time::now().toSec();
        current_angle = angular_speed * (t1 - t0); # rad/s
        velocity_publisher.publish(vel_msg);
        
        ros::spinOnce();
        loo_rate.sleep();
    }while(current_angle < relative_angle);
    vel_msg.angular.z = 0
    velocity_publisher.publish(vel_msg);
}
````
