### Frame assignment to robot manipulators

Frame assignment is quite important concept in robotics as we have to be able to descibe how a particular part of the robot is oriented with respect to any other part or object in this whole system. If you aren't quite familiari on what frames are and how they work, check out previous lessons on [Spatial descriptions and transformations](./__spatial_descriptions_and_transformations.ipynb).<br>
Before we go further let's look again at <code>**links and joints**</code> that make up a robot system. Any robot such as the one shown here can be thought of as a set of rigid bodies connected in a chain through multiple joints.

There are different types of joints that a robot can have such as screw and spherical type joints, but we will be focusing on a <code>**revolute type**</code>denoted <code>**R**</code> and <code>**prismatic type**</code> detoned by <code>**P**</code> joints.

<code>**revolute**</code> joint can rotate or revolve around its own axis while <code>**prismatic**</code> joint can move linearly up or down left or right. We will focus on these two joints as they are the simplest to work with due to single one degree of freedom they have, a.k.a<code>**DOF**</code><br>
A general purpose robot ussualy has about six DOF, more then six would be rendundant, and if less deficient. A good rule of thumb if robot has <code>**n**</code> of <code>**DOF**</code> it would have <code>**n**</code> joints and <code>**n-1**</code> of links.<br>
Let's have a look at how we can assign frames to each of the joints in robot system. There are six main conventions that can be used for assigning the frames to the joints of a robot.

## Conventions for fixing frames

<code>**1)**</code> Indetify the joint axes and imagine(or draw) a line along it's axis of movement<br>

Start with the <code>**base**</code> it is fixed and doesn't move. Then identify of what type joints system is composed. Imagine or draw a line throught the axis of movement for each of the joints. 

<code>**2)**</code> Assign the <code>**Z**</code> axis pointing along the joint axis

We should be able to assign or find the direction of these joint axes. For revolute joint the direction can be found using <code>**Right Hand Rule**</code> where the fingers bending direction represent positive rotation and the thumb would point to the positive direction of axis. Yet we don't know where are the origin of these frames are we just know the generic direction and line in which all three z axes are point in.

<code>**3)**</code>Set the origin of the frame of the i'th axis<br>

To asign origins to our three frames on our three joints.
Here we would consider
$${i}\rightarrow{i+1}$$ 
joint axis which means if we are to set the <code>**first axis**</code> we have to consider the <code>**second**</code>
$${1}\rightarrow{2}$$
$${2}\rightarrow{3}$$
And there are two main methods or conventions depending on how the axes are oriented:

 * Point of the interestion between joint axes <code>**(i and i + 1)**</code>. It means if the axes intesect with each other then the <code>**POINT OF INTERSECTION would denote the origin for the 1'st axis**</code>.<br>

----------------------------------------------
 
 *  Start of the common perpendicular between the joint axes <code>**(i and i+1)**</code>.It means if the axes are parallel to each other we would draw a common perpendicular between the two axes and the <code>**START of the common perpendicular would denote the origin for the 1'st axis**</code>.<br>

---------------------------------------------- 

<code>**4)**</code> Assigning x axis:
 * Along the common perpendicular
 * If the axes intersect, x <code>**would be normal vector to the intesecting planes of two axes**</code>

What if two axis intesect each other. Image two revolute joints connected throught some sort of link.
Both axes of revolutes joint would run through their axess of rotation. And they intersect in some point. There for to set the origin of first axis we would consider the second joint axis and origin of the first axis would be in that intesection point. In this case the Z of first axis would point up but it can point down as well depending on how the firs axis and second rotate.

<code>**NOTE**</code> X axis has been assigned opposite (against) to the rotation of <code>**second axis**</code>
