## Position vector

The first step in finding whereabouts in any robotic manipulator is to find its position in 3D space. Any position or a point is given with respect to world or universe coordinate system denoted by a letter in curly brackets <code>**{A}**</code>, <code>**{W}**</code>, etc.

The position of any robot links showed below or EOAT (end of arm tool) also called as end effector is given using position vector and is denoted by $$\begin{equation*}^Ap\end{equation*}$$. There can be multiple position vectors or even other coordinate systems defined for joints and all pisition vector and cooridante system would <code>**link back**</code> to the world or the universe cooridnate system. Due to that we use this notation:$$\begin{equation*}^Ap = \begin{bmatrix}
p_x \\
p_y \\
p_Z 
\end{bmatrix}\end{equation*}$$

<code>**A superscript**</code> shows the reference coordinate system and <code>**p**</code> that its a position vector.
If cartesian coordinates are used a position of point in space can be given through its <code>**x, y, z**</code> components

## Orientation Matrix

Along with the position the orientation of the manipulator must also be described. This shows how it is angled or oriented in space. In a case below a coordinate frame <code>**{B}**</code> was assignemt to robot end effector. It shows how the x, y, z axis are sets for this particular part of the robot. Using this the orientation of <code>**{B}**</code> can be express relative to any previous coordinate frame, in this case frame <code>**{A}**</code>. Numerically the orientation is expressed using <code>**3x3**</code> rotation matrix which comprises of stack of the unit vectors in the x, y ,z directions going from <code>**{B}**</code> to <code>**{A}**</code>.

\begin{equation*}
^A_B R = \begin{bmatrix} ^A\hat{X}_B & ^A\hat{Y}_B & ^A\hat{X}_B \end{bmatrix} = 
\begin{bmatrix}
r_{11} & r_{12} & r_{13} \\
r_{21} & r_{22} & r_{23} \\
r_{31} & r_{32} & r_{33}
\end{bmatrix} 
\end{equation*}

\begin{multline*}
\begin{split}
& ^A_B R  \\
\end{split}
\end{multline*} shows the orientation of <code>**{B}**</code> relative to <code>**{A}**</code>. <code>**Memorize this notation!**</code>

## Frames

<code>**Frame > Orientation matrix & Position vector**</code<

\begin{equation*}
\{B\} = 
\begin{bmatrix}
^A_B R  & ^A P_Borg \\
\end{bmatrix}
\end{equation*}

By combining orientation matrix with position vector we can fully describe whereabouts of any robotic manipulator this is knowed as a <code>**frame**</code>. Note that frame is just another coordiante system where the position of it is origin <code>**(notation org)**</code> and its orientation <code>**R**</code> are knowed with respect to previous coordinate system, in this case <code>**{A}**</code>. Analyzing this system of three frames <code>**{A}, {B}, {C}**</code> which looks quite similar to a simple robot with base one joint and a tool. Robot base corresponds to frame <code>**{A}**</code> which also would be the world coordinates, the joint would be frame <code>**{B}**</code> and the tool would be frame <code>**{C}**</code>. Through this we can see that we know frame B relative to frame A and we know frame C relative to B, but we do not know frame C relation to frame A, but however this can be calculated by using transformation matrix that we will go through a bit latter on. 

When we know how frames operate, we can map vectors $$\\vec{v}\$$ from one frame <code>**{B}**</code> to another <code>**{A}**</code> and when doing thise there are three main cases:<br>
1)<code>**Translation**</code><br>
2)<code>**Rotation**</code><br>
3))<code>**Combination of both**</code><br>

In case below we are looking into <code>**translation**</code>. A point is defined in frame <code>**{B}**</code> and has position vector <code>**Bp**</code>. 
Frame <code>**{B}**</code> has the same orientation as <code>**{A}**</code> it is only translated by some amount.

## Translated frames

So if we want to express this point <code>**P**</code> in term of frame <code>**{A}**</code> instead of frame <code>**{B}**</code> we calculate the following sum of <code>**two vectors**</code>:<br><br>
\begin{equation}
^AP=^BP + ^AP_Borg
\end{equation}

The point in <code>**{B}**</code> has not changed only its decription has changed.

## Rotated frames

Here frame <code>**{B}**</code> has been rotated with respect to frame <code>**{A}**</code>. The orientation of <code>**{B}**</code> differs from frame <code>**{A}**</code> yet there is not translation as the origin of <code>**{B}**</code> has not changed

A point in frame <code>**{B}**</code> can be represented in frame <code>**{A}**</code> by using 3x3 rotation matrix:
\begin{equation*}
^A_B R \end{equation*}<br>


\begin{equation*}
R_x = 
\begin{bmatrix}
1 & 0 & 0 \\
0 & \cos(\theta) & -sin(\theta) \\
0 & sin(\theta) & \cos(\theta)
\end{bmatrix},
R_y =
\begin{bmatrix}
\cos(\theta) &  0 &sin(\theta) \\
0 & 1 & 0 \\
-sin(\theta) & 0 & \cos(\theta)
\end{bmatrix}
R_z =
\begin{bmatrix}
\cos(\theta) & -sin(\theta) & 0 \\
sin(\theta) & \cos(\theta) & 0 \\
0 & 0 & 1
\end{bmatrix}
\end{equation*}

this would sum down to these three matrixes depending on how frame <code>**{B}**</code> was rotated in <code>**x**</code>,<code>**y**</code>,<code>**z**</code> directions for an angle of rotation: <br><br>\begin{equation*}\theta\end{equation*}:

The point <code>**P**</code> in frame <code>**{A}**</code> would be the rotation matrix times position vector \begin{equation*}
^B P \end{equation*}

\begin{equation*}
^AP= ^A_B R \times  ^BP \end{equation*}

### NOTE
This rotation matrix \begin{equation*}
^A_B R \end{equation*}<br>
can be a combination of those tree matrixes of any order, for example

\begin{equation*}
^A_B R = R_x R_y R_x
\end{equation*}


## Transformation frames

Frames can be translated and rotatated seperately but in a typical robotic system there are multiple translations and rotations about all three axis and this would properly alow us to move and orient a tool to it's desired location.

Frame <code>**{B}**</code> is rotated and translate by some amount with respect to frame <code>**{A}**</code>. By convention and ease of use we can represent rotation and translation in a single 4x4 <code>**transformation matrix**</code>

\begin{equation*}
^A_B T = 
\begin{bmatrix}
^A_B R  & ^A P_Borg \\
0  0  0 & 1 \\
\end{bmatrix}
\end{equation*}

This matrix is comprised of rotation matrix, position vector of translation and fourth row would have three zeros and a one to complete the matrix. In this scenario when we have a point <code>**P**</code> represented in frame <code>**{B}**</code> and we would like to express it in terms of frame <code>**{A}**</code>

\begin{equation*}
^AP= ^A_B T \times  ^BP \end{equation*}
