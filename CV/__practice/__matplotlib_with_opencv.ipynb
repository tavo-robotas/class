The first advantage that we can use from [simple and adaptive image thresholding](./__simple_and_adaptive_image_thresholding.ipynb) lesson is instead of using multiple windows to display different results of operation on image. So lets say you want to have it a single go and we can do this by using matplotlib 


```python
file_1 = '../images/gradient.jpg'
img_1  = cv.imread(file_1, 0);

_, th_binary     = cv.threshold(img_1, 127, 255, cv.THRESH_BINARY)
_, th_invers     = cv.threshold(img_1, 127, 255, cv.THRESH_BINARY_INV)
_, th_trunc      = cv.threshold(img_1, 127, 255,  cv.THRESH_TRUNC)
_, th_tozero     = cv.threshold(img_1, 127, 255, cv.THRESH_TOZERO)
_, th_tozero_inv = cv.threshold(img_1, 127, 255, cv.THRESH_TOZERO_INV)
```

Lets simply make two arrays that hold image titles in case of many examples and array for images

```python
titles = ['SOURCE', 'TH_BINARY', 'TH_BINARY_INV', 'TH_TRUNC', 'TH_TOZERO', 'TH_TOZERO_INV']
images = [img_1, th_binary, th_invers, th_trunc, th_tozero, th_tozero_inv ]

for i in range(len(images)):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
```
