# Hough Transform

## Basics

It's a popular technique to detect any shape if you can represent that shape in a mathematical form. It can detect the shape even if it is broken or distorted by a bit.

### Lines

A line in the image space be expressedd with two variables:
 * In the cartesian coordinate system: <code>**y = mx + c**</code>
 * In the polar coordinate system <code>**xcosθ + ysinθ = r**</code>

## Steps

 * Edge detection, e.g using the Canny edge detector;
 * Mapping of edge points to the Hough space and storage in an accumulator;
 * Interpretation of the accumulator to yield lines of infinte length. The intepretation os done by thresholding and posibbly other constrains;
 * Conversion of infinite lines to finite lines.
 
In opencv implements two kind of Hough line transforms
 * Standard Hough Transform (<code>**HoughLines**</code> method)
 * The Probabilistic Hough Line Transform (<code>**HoughLinesP**</code> method)
