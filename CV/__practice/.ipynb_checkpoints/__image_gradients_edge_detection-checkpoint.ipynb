### Image Gradients and Edge Detection

An image gradient is a directional change in the intesity or color inside an image. Image gradient is fundamental building block in image processing. We often use these gradients to find the edges inside image. There are several method available in opencv library. All of them are different gradient functions which uses different mathematical operations to produce the required image. While Laplaciant calculates derivatives where as Sobel method is joint Gaussian and differentiation operations.

```python
import cv2 as  cv
import numpy as np
from matplotlib import pyplot as plt

file = '../images/messi.jpg'
source = cv.imread(file, 0)

#Laplacian derivatives
# CV_64F 64bit float data type, that support negative numbers
# ksize=1 is default optional argument
lap = cv.Laplacian(source, cv.CV_64F, ksize=1)
lap = np.uint8(np.absolute(lap))

#Sobel
# 1 means use x for dx
# 0 means use y for dy
# ksize=1 is default optional argument

sobelx = cv.Sobel(source,cv.CV_64F, 1, 0)
sobelx = np.uint8(np.absolute(sobelx))
sobely = cv.Sobel(source,cv.CV_64F, 0, 1)
sobely = np.uint8(np.absolute(sobely))

titles = ['source', 'laplacian', 'sobelx', 'sobely']
images = [source, lap, sobelx, sobely]

plt.subplots(figsize=(20,20))
for i in range(len(images)):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
