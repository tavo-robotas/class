# Edge Detection

## Image Gradients and Edge Detection

An image gradient is a directional change in the intesity or color inside an image. Image gradient is fundamental building block in image processing. We often use these gradients to find the edges inside image. There are several method available in opencv library. All of them are different gradient functions which uses different mathematical operations to produce the required image. While Laplaciant calculates derivatives where as Sobel method is joint Gaussian and differentiation operations. 

### Laplacian derivatives


### Sobel X
the change in direction in the intesity is in x direction

### Sobel Y
the change in direction in the intesity is in y direction


```python
import cv2 as  cv
import numpy as np
from matplotlib import pyplot as plt

file = '../images/sudoku.jpg'
source = cv.imread(file, 0)

#Laplacian derivatives
# CV_64F 64bit float data type, that support negative numbers
# ksize=1 is default optional argument
lap = cv.Laplacian(source, cv.CV_64F, ksize=1)
lap = np.uint8(np.absolute(lap))

#Sobel
# 1 means use x for dx
# 0 means use y for dy
# ksize=1 is default optional argument

sobelx = cv.Sobel(source,cv.CV_64F, 1, 0)
sobelx = np.uint8(np.absolute(sobelx))
sobely = cv.Sobel(source,cv.CV_64F, 0, 1)
sobely = np.uint8(np.absolute(sobely))

sobel_combined = cv.bitwise_or(sobelx, sobely)

titles = ['source', 'laplacian', 'sobelx', 'sobely', 'sobel_combined']
images = [source, lap, sobelx, sobely, sobel_combined]

plt.subplots(figsize=(20,20))
for i in range(len(images)):
    plt.subplot(3,2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```

## Canny edge detection 

Its an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images, composed of 5 steps:<br>
 * Noise reduction, applying Gaussian filter to smooth the image in order to remove noise;
 * Gradient calculation to find intesity gradients of image;
 * Non-maximum supression to get rid of spurious response to edge detection;
 * Double threshold to determine the potential edges;
 * Edge tracking by hysteresis and track edges by suppressing all the other that are weak or not connected to strong edges. 

```python
import numpy as np
from matplotlib import pyplot as plt

file = '../images/sudoku.jpg'
source = cv.imread(file, 0)

# previous edge detection methods
lap = cv.Laplacian(source, cv.CV_64F, ksize=1)
lap = np.uint8(np.absolute(lap))
sobelx = cv.Sobel(source,cv.CV_64F, 1, 0)
sobelx = np.uint8(np.absolute(sobelx))
sobely = cv.Sobel(source,cv.CV_64F, 0, 1)
sobely = np.uint8(np.absolute(sobely))
sobel_combined = cv.bitwise_or(sobelx, sobely)

# Canny edge detection method
canny = cv.Canny(source, 100, 200)

titles = ['source', 'laplacian', 'sobelx', 'sobely', 'sobel_combined', 'canny']
images = [source, lap, sobelx, sobely, sobel_combined, canny]

plt.subplots(figsize=(20,20))
for i in range(len(images)):
    plt.subplot(3,2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show() 
```
