## Prime task of computer vision

Efective machine interaction with enviroment over image and video. This means that computers should make <code>**inferences**</code> about images without any human help. Computers may conduct tasks such as determining if an object is in a photo, how many objects are in the photo, where these objects are located in the photo, and much more. Although this seems very simple because humans can effortlessly view the world around them; however, teaching a computer to see like a human to be honest is <code>**extremly difficult because we still dont really understand how human vision works**</code> But we are getting there, and the good news are the more of us  work in computer vision research the more progress we collectivly achieve.

## Origins and milestones

### Year 1959. Ocular dominance columns
The historical roots of computer vision began in the <code>**1950's**</code> and is closely intertwined with the study of human vision. To teach computers how to be able to "see," researchers have sought inspiration from visual perception in humans. <code>**Humbel and Wiesel**</code> conducted a research on sensory processing. They were able to show that visual stimuli travels toward the visual cortex in the braind and the signals thren branch out to [ocular dominance columns](https://www.sciencedirect.com/science/article/abs/pii/0006899374904168)
Individaul cells are triggered by particular visual details such as <code>**lines**</code> or <code>**edges**</code> in particular orientations. The information is then passed to other part of brain to stimulate other group of cells. This idea of <code>**hierarchical processing**</code> can be senn in many modern computer vision algorithms such as <code>**neural network**</code>

### Year 1963. Block world
This concept was a precursor to computer vision. It was introduced by American scientist <code>**Larry Roberts**</code>. Roberts described how the distinc edges of blocks helped humans understand the shape of the block regadless of its orientation or color intesity. The idea of <code>**detecting edges and corners**</code> would later be widely used in computer vision algorithms.

### Year 1966. MIT Summer vision project

Ironically, computer vision was first introduced as an MIT summer project. The summer workers had to work efficiently to "participate in the constrcution of a system complex enough to be a real landmark in the development of <code>**patter recognition**</code>. 50 year later, today, scientist are still trying to fully understand pattern recognition and computer vision.

### Year 1970. Geons
In the 1970's, vision scientist <code>**Irving Biederman**</code> and cognitive scientist <code>**Molly Potter**</code> both discovered how efficiently humans are able to recognize and interpret what they see.<br>
Biederman also proposed the <code>**Recognition by components theory**</code>, which suggest that humans are able to recognize objects by splitting them into their main component parts, called <code>**geons**</code>. He would go on to study face recognition as well and argue that it is a process distinct from recognition of objects.<br>
Ms Potter determined whether or not a sequance of pictures, displayed at rates approximate to those of eye fixations are undestood and then forgotten. Potter conculded that a scene is processed within around <code>**100 ms**</code> but humans need <code>**400 ms**</code> to ensure that conceptual masking does not occur due to a sebsequent picture.

### Year 1986. CNN

<code>**Yann LeCun**</code> at AT&T Bell Laboratories developed convoliutional neural networks that were successfully used for <code>**zip code recognition**</code> by the postal service. This was the beginning of the CNN's that are widely used in the computer vision today.

### Year 1996. Speed of processing

<code>**Simon Thorpe**</code> authored the [Speed of processing in the human visual system](https://pubmed.ncbi.nlm.nih.gov/8632824). Like Biederman and Potter, he also noted how humans can instantaneously recognize well known or commonly seen objects and scenes. He set out to determine exactly how <code>**instantaneous**</code> human eye processing is by using event-related potentials <code>**ERP's**</code>, or measured brain activity in response to a certain stimuli, which reveal signs of neural processing well before a physical response is elicited.<br>
Thorpe devised a <code>**go/no-go**</code> categorization task, where subjects decided whether a photograph, flashed for 20 ms, contained an animal. Results showed that frontal negativity for no-go trials developed roughly 150 ms after the image was displayed.<br>
Thorpe concluded that visual processing can be reached in under 150 ms.

### Year 1996. Normalized cut

<code>**Jianbo Shi and Jitendra Malik**</code> normalized cuts algorithm work  was a milestone work in computer vision researchers starting to use color images and portrayal of the real world. In this work that tackled <code>**image segmentation**</code>, the driving question was the ability to separate an image into distinct and sensible parts. It is essential for a computer to be able to <code>**recognize and label the segmented parts**</code> of an image before combining them and analyzing what is going on in the picture. 


### Year 2009. ImageNET

<code>**Prof. Fei-Fei Li and her colleagues**</code> put together a [massive public database](http://image-net.org/) of images, over <code>**22 thousand** categories</code> with <code>**14 million images**</code>. The purpose was to provide public data to ease the concern of the need for more data when working on <code>**object recognition**</code>. ImageNet hold annual competitions and challenges for object recognition. 

### Year 2012. AlexNET

In 2012, graduate student [Alex Krizhevsky](http://papers.nips.cc/paper/4824-imagenet-classification-with-deep-convolutional-neural-networks.pdf) in Geoffrey Hintons research group at the University of Toronto trained convolutional neural networks on the large scale ImageNet dataset and achieved image classification performance that dramatically beat previous state-of-the-art methods. This created a surge of work and interest in <code>**deep learning methods**</code> using neural networks.

### Simple example for some fun

```python
## use python pip to install required libraries
## pip3 install opencv-python
## pip3 install imutils

import cv2
import imutils
cap = cv2.VideoCapture(0)
lower = (100, 130, 70) # HSV lower values
upper = (160, 255, 220) # HSV upper values
while cap.isOpened():
    retain, frame = cap.read()
    blurred = cv2.GaussianBlur(frame, (11, 11), 0)
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, lower, upper)
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)
    outlines = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = imutils.grab_contours(outlines)
    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        ((x, y), radius) = cv2.minEnclosingCircle(c)
        M = cv2.moments(c)
        center = (int(M['m10'] / M['m00']), int(M['m01'] / M['m00']))
        if radius > 10:
            cv2.circle(frame, (int(x), int(y)), int(radius), (0, 255, 255), 2)
            cv2.circle(frame, center, 5, (255, 255, 255), -1)
    cv2.imshow('image', frame)
    cv2.imshow('mask', mask)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
```
